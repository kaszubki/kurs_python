```python
# -*- coding: utf-8 -*-

import pandas as pa
import numpy as np
import matplotlib.pyplot as plt
import unicodedata
%matplotlib inline
```

```python
data = pa.read_csv('wybory-2015-obwody.csv')
```

```python
data.describe()
```

```python
nrow = len(data)
print nrow
```

```python
colnames = list(data.columns)
```

```python
for i in xrange(len(colnames)):
    print "{0} : {1}".format(i, colnames[i])
```

```python
data.fillna(0, inplace=True)
```

```python
adresy = data["Adres"]
```

```python
moj_adres = [i for i in xrange(len(adresy)) if (adresy[i].find("Warszawa") != -1) and (adresy[i].find("Hirszfelda") != -1)]
print moj_adres
```

```python
small_data = data.ix[moj_adres]
print small_data
```

```python
glosy = []
komitety = []
for id_partii in xrange(28, 45):
    komitet = colnames[id_partii]
    ile = sum(int(small_data[colnames[id_partii]][moj_adres[i]]) for i in xrange(len(moj_adres)))
    glosy.append(ile)
    komitety.append(komitet)
    print "Liczba glosow na komitet {0} wyniosła {1}".format(komitet, ile)

    

komitety = [unicode(repr(k)) for k in komitety]

#for k in komitety:
#    unicodedata.normalize('NFKD', k).encode('ascii','ignore')

    
print glosy
print komitety
    
OX = np.arange(len(glosy))
plt.bar(range(len(glosy)), glosy)
plt.xticks(np.array(range(len(glosy))) + 0.5, komitety, rotation="vertical")
plt.show()
```

```python
komitety2 = [komitety[i] for i in xrange(len(glosy)) if glosy[i] != 0]
glosy2 = [glosy[i] for i in xrange(len(glosy)) if glosy[i] != 0]

explode = [0.2] * len(glosy2)
plt.pie(glosy2, labels=range(1, len(glosy2) + 1), explode=explode, autopct='%1.1f%%', shadow=True, startangle=30)
plt.legend(komitety2)
plt.show()
```
